```python
class Optimizer:
    def __init__(self, params):
        self.params = list(params)

    def zero_grad(self):
        for param in self.params:
            if param.grad is not None:
                param.grad.detach_()
                param.grad.zero_()
    
    def pre_step(self):
        pass
    
    def step(self):
        pass
    
    def update_param(self, p):
        raise NotImplementedError()
```

```python
def optimize_function(fn, optim, optim_args, start_point, num_iter = 50):
    weigths = nn.Parameter(torch.FloatTensor(start_point), requires_grad=True)

    optim = optim(params=[weigths], **optim_args)
    points = []
    losses = []
    for i in range(num_iter):
        if hasattr(optim, 'pre_step'):
            optim.pre_step()
        loss = fn(weigths[0], weigths[1])
        points.append(weigths.data.detach().clone())
        losses.append(loss.item())
        loss.backward()
        optim.step()
        optim.zero_grad()
    points = torch.stack(points, axis=0)
    losses = torch.FloatTensor(losses)
    return points, losses
```

```python
def compare_optimizers(
    fn,
    optim_list,
    start_point,
    x_range=(-5, 5),
    y_range=(-5, 5),
    xstep=0.2,
    ystep=0.2,
    minima=None,
    num_iter = 50,
):
    xmin, xmax = x_range
    ymin, ymax = y_range
    x, y = np.meshgrid(np.arange(xmin, xmax + xstep, xstep), np.arange(ymin, ymax + ystep, ystep))
    z = fn(torch.from_numpy(x), torch.from_numpy(y))
    z = z.detach().numpy()
    fig, ax = plt.subplots(figsize=(10, 6))

    ax.contour(x, y, z, levels=np.logspace(0, 5, 35), norm=LogNorm(), cmap=plt.cm.jet)
    if minima:
        ax.plot(*minima, 'r*', markersize=18)

    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')

    ax.set_xlim((xmin, xmax))
    ax.set_ylim((ymin, ymax))
    
    colors = cm.rainbow(np.linspace(0, 1, len(optim_list)))
    
#     кривая лоса
    name_losses = {}
    
    for c, (name, optim, args) in zip(colors, optim_list):
        points, losses = optimize_function(fn, optim, args, start_point, num_iter)
        ax.quiver(
            points[:-1, 0], points[:-1, 1],  
            points[1:, 0] - points[:-1, 0], points[1:, 1] - points[:-1, 1], 
            scale_units='xy', angles='xy', scale=1, color=c,
            label=name
        )
        name_losses[name] = losses
    
    
    ax.legend()
    plt.show()
    
#     отрисовка лоса
    plt.figure()
    x = np.array(np.arange(num_iter))
    for name, losses in name_losses.items():
        plt.plot(x, losses.numpy(), label=name)

    plt.legend(
        loc='upper right',
        fontsize=8
    )
```

### Оптимизаторы SGD

```python
class SGD(Optimizer):
    def __init__(self, params, lr=1e-2):
        super().__init__(params)
        self.lr = lr
        
    def step(self):
        with torch.no_grad():
            for p in self.params:
                if p.grad is None:
                    continue
                p -= self.lr * p.grad
        
```

```python
class Momentum(Optimizer):
    def __init__(self, params, lr=1e-2, gamma=0.9):
        super().__init__(params)
        self.lr = lr
        self.v = [torch.zeros_like(p) for p in self.params]
        self.gamma = gamma
        
    def step(self):
        with torch.no_grad():
            for v, p in zip(self.v, self.params):
                if p.grad is None:
                    continue
                v.copy_(self.gamma * v + self.lr * p.grad)
                p -= self.lr * v
```

```python
class RMSProp(Optimizer):
    def __init__(self, params, lr=1e-2, gamma=0.9, eps=1e-9):
        super().__init__(params)
        self.lr = lr
        self.G = [torch.zeros_like(p) for p in self.params]
        self.gamma = gamma
        self.eps = eps
        
    def step(self):
        with torch.no_grad():
            for g, p in zip(self.G, self.params):
                if p.grad is None:
                    continue
                g.copy_(self.gamma * g + (1 - self.gamma) * p.grad**2)
                p -= self.lr / torch.sqrt(g + self.eps) * p.grad
```

```python
class Adagrad(Optimizer):
    def __init__(self, params, lr=1e-2, gamma=0.9, eps=1e-9):
        super().__init__(params)
        self.lr = lr
        self.G = [torch.zeros_like(p) for p in self.params]
        self.gamma = gamma
        self.eps = eps
        
    def step(self):
        with torch.no_grad():
            for g, p in zip(self.G, self.params):
                if p.grad is None:
                    continue
                g.copy_((p.grad**2).sum())
                p -= self.lr / torch.sqrt(g + self.eps) * p.grad
```

```python
class Adadelta(Optimizer):
    def __init__(self, params, lr, gamma=0.9, eps=1e-9):
        super().__init__(params)
        self.G = [torch.zeros_like(p) for p in self.params]
        self.delta = [torch.zeros_like(p) + 1e-5 for p in self.params]
        self.gamma = gamma
        self.eps = eps
        
    def step(self):
        with torch.no_grad():
            for g, delta, p in zip(self.G, self.delta, self.params):
                if p.grad is None:
                    continue
                g.copy_(self.gamma * g + (1 - self.gamma) * p.grad**2)
                numerator = torch.sqrt(delta + self.eps) / torch.sqrt(g + self.eps) * p.grad
                p -= numerator
                delta.copy_(self.gamma * delta + (1 - self.gamma) * numerator**2)
```

```python
class NAG(Optimizer):
    def __init__(self, params, lr=1e-2, gamma=0.9):
        super().__init__(params)
        self.lr = lr
        self.v = [torch.zeros_like(p) for p in self.params]
        self.gamma = gamma
        
    def pre_step(self):
        with torch.no_grad():
            for v, p in zip(self.v, self.params):
                p -= self.gamma * v
        
    def step(self):
        with torch.no_grad():
            for v, p in zip(self.v, self.params):
                if p.grad is None:
                    continue
                p += self.gamma * v
                v.copy_(self.gamma * v + self.lr * p.grad)
                p -= v
```

```python
class Adam(Optimizer):
    def __init__(self, params, lr=1e-2, beta1=0.9, beta2=0.999, eps=1e-8):
        super().__init__(params)
        self.lr = lr
        self.beta1=beta1
        self.beta2=beta2
        self.t = 0

        self.m = [torch.zeros_like(p) for p in self.params]
        self.v = [torch.zeros_like(p) for p in self.params]
        
        self.eps = eps
        
#     def pre_step(self):
#         with torch.no_grad():
#             for m, v, p in zip(self.m, self.v, self.params):
#                 self.pre_step = self.lr * (m / (1 - self.beta1)) / (torch.sqrt(v / (1 - self.beta2)) + self.eps)
#                 p -= self.pre_step

#                 m.copy_(self.beta1 * m + (1 - self.beta1) * p.grad)
#                 v.copy_(self.beta1 * v + (1 - self.beta2) * p.grad**2)
        
                

    def step(self):
        with torch.no_grad():
            for m, v, p in zip(self.m, self.v, self.params):
                if p.grad is None:
                    continue
#                 p += self.pre_step
                
                m.copy_(self.beta1 * m + (1 - self.beta1) * p.grad)
                v.copy_(self.beta1 * v + (1 - self.beta2) * p.grad**2)
                
                p -= self.lr * m / (1 - self.beta1**self.t) / (torch.sqrt(v / (1 - self.beta2**self.t)) + self.eps)
        self.t += 1
```

### Обучение
В качестве нейронной сети: линейные слои, нелинейность=tanh.
Одним из переменных параметров является оптимизатор SGD

```python
def train(network, epochs, learning_rate, optim):
    loss = nn.NLLLoss()
    optimizer = optim(network.parameters(), lr=learning_rate)
    train_loss_epochs = []
    test_loss_epochs = []
    train_accuracy_epochs = []
    test_accuracy_epochs = []
    try:
        for epoch in range(epochs):
            losses = []
            accuracies = []
            for X, y in train_loader:
                network.zero_grad()
                prediction = network(X)
                loss_batch = loss(prediction, y)
                losses.append(loss_batch.item())
                loss_batch.backward()
                optimizer.step()
                accuracies.append((np.argmax(prediction.data.numpy(), 1)==y.data.numpy()).mean())
            train_loss_epochs.append(np.mean(losses))
            train_accuracy_epochs.append(np.mean(accuracies))
            
            losses = []
            accuracies = []    
            for X, y in test_loader:
                prediction = network(X)
                loss_batch = loss(prediction, y)
                losses.append(loss_batch.item())
                accuracies.append((np.argmax(prediction.data.numpy(), 1)==y.data.numpy()).mean())
            test_loss_epochs.append(np.mean(losses))
            test_accuracy_epochs.append(np.mean(accuracies))
            
            clear_output(True)
            print('\rEpoch {0}... (Train/Test) NLL: {1:.3f}/{2:.3f}\tAccuracy: {3:.3f}/{4:.3f}'.format(
                        epoch, train_loss_epochs[-1], test_loss_epochs[-1],
                        train_accuracy_epochs[-1], test_accuracy_epochs[-1]))
            plt.figure(figsize=(12, 5))
            plt.subplot(1, 2, 1)
            plt.plot(train_loss_epochs, label='Train')
            plt.plot(test_loss_epochs, label='Test')
            plt.xlabel('Epochs', fontsize=16)
            plt.ylabel('Loss', fontsize=16)
            plt.legend(loc=0, fontsize=16)
            plt.grid()
            plt.subplot(1, 2, 2)
            plt.plot(train_accuracy_epochs, label='Train accuracy')
            plt.plot(test_accuracy_epochs, label='Test accuracy')
            plt.xlabel('Epochs', fontsize=16)
            plt.ylabel('Loss', fontsize=16)
            plt.legend(loc=0, fontsize=16)
            plt.grid()
            plt.show()
    except KeyboardInterrupt:
        pass
    return train_loss_epochs, \
           test_loss_epochs, \
           train_accuracy_epochs, \
           test_accuracy_epochs
```

#### Линейная нейронная сеть

```python
image_size = 28
class LinearClassifier(nn.Module):
    def __init__(self, image_size):
        super(LinearClassifier, self).__init__()

        self.linear_layers1 = nn.Sequential(nn.Linear(784, 100), nn.Tanh())
        self.linear_layers2 = nn.Sequential(nn.Linear(100, 100), nn.Tanh())
#         self.linear_layers3 = nn.Sequential(nn.Linear(500, 100), nn.Sigmoid())
        self.linear_layers_res = nn.Sequential(nn.Linear(100, 10), nn.LogSoftmax(dim=1))
    
    def forward(self, x):

        x = x.view(x.size(0), -1)
        x_linear = torch.zeros_like(x)
        x_linear.copy_(x)
        
        x_linear = self.linear_layers1(x_linear)
        x_linear = self.linear_layers2(x_linear)
#         x_linear = self.linear_layers3(x_linear)
        x_linear = self.linear_layers_res(x_linear)
        
        return x_linear

network_linear = LinearClassifier(image_size=28)
```

```python
# SGD
tr_l_lin_sgd, ts_l_lin_sgd, tr_ac_lin_sgd, ts_ac_lin_sgd = train(network_linear, 5, 1e-3, SGD)
```

```python
# Adagrad
tr_l_lin_adagrad, ts_l_lin_adagrad, tr_ac_lin_adagrad, ts_ac_lin_adagrad = train(network_linear, 5, 1e-3, Adagrad)
```

```python
# AdaDelta
tr_l_lin_adadelta, ts_l_lin_adadelta, tr_ac_lin_adadelta, ts_ac_lin_adadelta = train(network_linear, 5, 1e-2, Adadelta)
```

```python
# Adam
tr_l_lin_adam, ts_l_lin_adam, tr_ac_lin_adam, ts_ac_lin_adam = train(network_linear, 5, 1e-5, torchAdam)
```

#### Сверточная нейронная сеть

```python
image_size = 28
channels = 1
class ConvClassifier(nn.Module):
    def __init__(self, image_size):
        super(ConvClassifier, self).__init__()
        self.conv_layers = nn.Sequential(nn.Conv2d(channels, 16, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2))
#         self.conv_layers = nn.Sequential(nn.Conv2d(16, 24, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2))
        self.linear_layers = nn.Sequential(nn.Linear(3136, 10), nn.LogSoftmax(dim=1))
    
    def forward(self, x):
        x_conv = torch.zeros_like(x)
        x_conv.copy_(x)
        
        
        x_conv = self.conv_layers(x_conv)
        
        x_conv = x_conv.view(x.size(0), -1)

        x_conv = self.linear_layers(x_conv)

        return x_conv
```

```python
# Adam
network_conv = ConvClassifier(image_size=28)
tr_l_conv_adam, ts_l_conv_adam, tr_ac_conv_adam, ts_ac_conv_adam = train(network_conv, 5, 1e-2, torchAdam)
```

```python
# SGD
network_conv = ConvClassifier(image_size=28)
tr_l_conv_sgd, ts_l_conv_sgd, tr_ac_conv_sgd, ts_ac_conv_sgd = train(network_conv, 5, 1e-2, Momentum)
```

```python
# Adagrad
network_conv = ConvClassifier(image_size=28)
tr_l_conv_adagrad, ts_l_conv_adagrad, tr_ac_conv_adagrad, ts_ac_conv_adagrad = train(network_conv, 5, 1e-2, Adagrad)
```
